```python
# import dependencies
import numpy as np
import pandas as pd
from pathlib import Path
from sklearn.metrics import confusion_matrix, classification_report
```

Load in data

```python
# Read in CSV to df
file_path = Path("Resources/test.csv")
df_test = pd.read_csv(file_path)

file_path = Path("Resources/train.csv")
df_train = pd.read_csv(file_path)

df_train
```

General Cleaning

```python
# dtm # of NaN values in each column
list_of_columns = ['ID', 'Customer_ID', 'Month', 'Name', 'Age', 'SSN', 'Occupation', 'Annual_Income', 'Monthly_Inhand_Salary',
                   'Num_Bank_Accounts', 'Num_Credit_Card', 'Interest_Rate', 'Num_of_Loan', 'Type_of_Loan', 'Delay_from_due_date',
                   'Num_of_Delayed_Payment', 'Changed_Credit_Limit', 'Num_Credit_Inquiries', 'Credit_Mix', 'Outstanding_Debt',
                   'Credit_Utilization_Ratio', 'Credit_History_Age', 'Payment_of_Min_Amount', 'Total_EMI_per_month', 'Amount_invested_monthly',
                   'Payment_Behaviour', 'Monthly_Balance', 'Credit_Score']


nan_counts = df_train[list_of_columns].isna().sum()
nan_counts
```

```python
# drop duplicate or unnecessary info for classsification model
# Name and SSN are duplicate due to Customer_ID
columns_to_drop = ['Month', 'Name', 'SSN', 'Monthly_Inhand_Salary', 'Type_of_Loan']
df_train.drop(columns=columns_to_drop, inplace=True)

df_train
```

```python
# drop any row with null value NaN
df_train.dropna(inplace=True)
df_train
```

```python
# clean Credit_Utilization_Ratio

# cast as int & round
df_train["Credit_Utilization_Ratio"] = df_train["Credit_Utilization_Ratio"].astype(float)
df_train["Credit_Utilization_Ratio"] = df_train["Credit_Utilization_Ratio"].round(2)

df_train.head(3)
```

```python
# convert credit history to numerical

# Function to convert age string to total years
def convert_age_to_years(age_str):
    years, months = map(str.strip, age_str.split('Years and'))
    total_months = int(years) * 12 + int(months.split()[0])
    total_years = total_months / 12  # Convert total months to years
    return round(total_years, 1)  # Round to 1 decimal place

# Convert age strings to total years
df_train['Credit_History_Age'] = df_train['Credit_History_Age'].apply(convert_age_to_years).astype(float)

df_train.head(3)
```

```python
df_train
```

Use Regular Expressions to Remove Typos

```python
columns_to_regex = ['Annual_Income', 'Age', 'Delay_from_due_date', 'Num_of_Loan','Amount_invested_monthly', 'Monthly_Balance']

for column in columns_to_regex:
    df_train[column] = df_train[column].astype(str).str.replace('_', '')
    df_train[column] = df_train[column].astype(float)  # Convert the column back to float

df_train['Age'] = df_train['Age'].astype(int)

df_train
```

```python
df_train.info()
```

```python
df_train['Changed_Credit_Limit'] = df_train['Changed_Credit_Limit'].replace('-', '')
df_train['Changed_Credit_Limit'] = pd.to_numeric(df_train['Changed_Credit_Limit'], errors='coerce')
df_train['Changed_Credit_Limit'] = df_train['Changed_Credit_Limit'].fillna(0) 
```

```python
# List of columns to round and cast as floats
columns_to_round = ['Num_of_Loan', 'Delay_from_due_date', 'Amount_invested_monthly', 'Monthly_Balance', 'Changed_Credit_Limit']

# Round the columns to 2 decimal places and cast them as floats
df_train[columns_to_round] = df_train[columns_to_round].round(2).astype(float)

df_train
```
